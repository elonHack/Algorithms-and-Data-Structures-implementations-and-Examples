#Implementing the document distance problem

```python
from os.path import splitdrive
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

d1 = "In the quiet town of Willowbrook,  by an elusive figure known as the \"Forest Guardian.\" Many had ventured into the woods, driven by the allure of untold riches, but none had returned with anything more than fanciful stories. As autumn leaves painted the landscape with hues of gold and crimson, a new group of treasure seekers gathered, determined to decipher the enigmatic clues and unlock the secrets hidden within the heart of Willowbrook's forest."

d2 = "Among the eager adventurers was Sarah, a young and fearless explorer with a heart full of curiosity. Her grandfather had been one of the countless seekers who had disappeared into the woods, leaving behind a worn map and a tale of elusive wonders. Sarah believed that her destiny was intertwined with the forest's mysteries. She gathered a diverse band of companions, each with their own unique skills and stories, and set off on a journey that would test their wit, courage, and friendship. As the morning sun filtered through the towering trees, casting long shadows on the forest floor, Sarah and her newfound allies ventured forth, ready to confront the challenges and unravel the legend of the Forest Guardian"

def split_into_dict(doc):
  total_dict = {}
  splited = doc.split(" ")
  for word in splited:
    if word in total_dict.keys():
      total_dict[word] += 1
    else:
      total_dict[word] = 1
  return total_dict




```

```python
#Generate the graphs
grap1 = split_into_dict(d1)
grap2 = split_into_dict(d2)

print(grap1)
print(grap2)
```

```python

# Create line plots for the values
plt.figure(figsize=(15, 5))

# Plot for d1

plt.plot(grap1.keys(), grap1.values(), marker='o', linestyle='-', color='b')
plt.xlabel('Keys')
plt.ylabel('Values')
plt.title('Line Plot for d1')
plt.xticks(rotation=90)
plt.show()
```

```python
# Create line plots for the values
plt.figure(figsize=(15, 5))

# Plot for d2

plt.plot(grap2.keys(), grap2.values(), marker='o', linestyle='-', color='r')
plt.xlabel('Keys')
plt.ylabel('Values')
plt.title('Line Plot for d2')
plt.xticks(rotation=90)
plt.show()
```
